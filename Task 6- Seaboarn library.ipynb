# Seaboarn Library

Seaborn is a library that uses Matplotlib underneath to plot graphs.  
It will be used to visualize random distributions.  
Import the Seaborn module in your code using the following statement:   
import seaborn as sns

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5])
plt.show()

#Plotting a Distplot Without the Histogram

import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)
plt.show()
```

# Scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")
# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
# Add titles and labels
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
# Show the plot
plt.show()
```

# Histogram

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

#  Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()
```

#  Line Plot

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

# Pair plot

```python
sns.pairplot(data=tips, hue="sex")
plt.show()
```

# Box Plot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

# Regression plot

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.title("Regression Plot of Total Bill vs. Tip (Faceted by Smoker)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
